```python
import numpy as np
import pandas as pd

score=np.random.randint(50,101,size=(50,5))
score_df=pd.DataFrame(score,
             columns=['國文','英文','數學','地理','歷史'],
             index=range(1,51))
score_df.index.name='學號'
score_df.columns.name='科目'
sum_series=score_df.sum(axis=1)
mean_series=score_df.mean(axis=1)
sum_series.rank(method='min',ascending=False)
score_df
```

```python
sum_series=score_df.sum(axis=1)
mean_series=score_df.mean(axis=1)
rank_series=sum_series.rank(method='min',ascending=False)
score_df['總分'] = sum_series
score_df['平均'] = mean_series
score_df['排名'] = rank_series
score_df
```

```python
score_sorted_df = score_df.sort_values(by='排名',ascending=True)
score_sorted_df
```
