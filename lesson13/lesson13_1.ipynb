```python
#標準寫法
import matplotlib.pyplot as plt
#1.製作figure
fig=plt.figure()
#2.在1.的figure製作一個axes
ax_1=fig.add_subplot(1,1,1)
#3.將圖的數據存在axes中
X=[0,1,2]
Y=[3,4,5]
ax_1.plot(X,Y,'bo-.')
#4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
figure=plt.figure()
ax_1=figure.add_subplot(1,3,1)
ax_2=figure.add_subplot(1,3,2)
ax_3=figure.add_subplot(1,3,3)

ax_1.set_title("1")
ax_2.set_title("2")
ax_3.set_title("3")

plt.show()
```

```python
import matplotlib.pyplot as pit
student1:dict[str,int]={'chinese':78,'en':80,'math':90,'hist':70,'discover':99}
student2:dict[str,int]={'chinese':77,'en':88,'math':92,'hist':71,'discover':97}
subject=list(student1.keys())
sc1=list(student1.values())
sc2=list(student2.values())

figure=plt.figure(figsize=(10,5))
axes=figure.add_subplot(1,1,1)
axes.plot(subject,sc1)
axes.plot(subject,sc2)
plt.show()
```
